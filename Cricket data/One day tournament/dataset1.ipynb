```python
# ==========================================
# Cricket One Day Data Analysis
# Univariate, Bivariate & Multivariate
# Author: Yash Parmar
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Load Dataset
# -------------------------------
df = pd.read_csv("/content/one_day.csv")   # change path if needed

# Convert Required Columns
columns_to_convert = ['Runs', 'Ave', 'SR', '100', '50', 'Mat', 'Inns']
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna()

# =====================================================
#  1 UNIVARIATE ANALYSIS
# =====================================================

# Distribution of Runs
plt.figure()
plt.hist(df['Runs'], bins=15)
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()

# Distribution of Strike Rate
plt.figure()
plt.hist(df['SR'], bins=15)
plt.title("Distribution of Strike Rate")
plt.xlabel("Strike Rate")
plt.ylabel("Frequency")
plt.show()

# Top 10 Players by Runs (Bar Chart)
top_runs = df.sort_values(by='Runs', ascending=False).head(10)

plt.figure()
plt.bar(top_runs['Player'], top_runs['Runs'])
plt.xticks(rotation=45)
plt.title("Top 10 Players by Runs")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.tight_layout()
plt.show()


# =====================================================
#  2 BIVARIATE ANALYSIS
# =====================================================

# Batting Average vs Strike Rate
plt.figure()
plt.scatter(df['Ave'], df['SR'])
plt.title("Batting Average vs Strike Rate")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.show()

# Matches vs Runs
plt.figure()
plt.scatter(df['Mat'], df['Runs'])
plt.title("Matches vs Runs")
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.show()


# =====================================================
#  3 MULTIVARIATE ANALYSIS
# =====================================================

# Runs vs Strike Rate colored by Average
plt.figure()
scatter = plt.scatter(df['Runs'], df['SR'], c=df['Ave'])
plt.colorbar(scatter)
plt.title("Runs vs Strike Rate (Colored by Average)")
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.show()

# Runs vs Matches sized by Centuries
plt.figure()
plt.scatter(df['Mat'], df['Runs'], s=df['100']*20)
plt.title("Matches vs Runs (Bubble Size = Centuries)")
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.show()

print("Univariate, Bivariate and Multivariate Analysis Completed ")
```
