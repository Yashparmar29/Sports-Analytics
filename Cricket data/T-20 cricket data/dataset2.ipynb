```python
# ==========================================
# Complete T20 Match Data Visualization
# Author: Yash Parmar
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Load Cleaned Dataset
# -------------------------------
df = pd.read_csv("/content/Twenty_20_data.csv")

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna()

# Extract Year
df['Year'] = df['Date'].dt.year

# =====================================================
#  1 UNIVARIATE ANALYSIS
# =====================================================

# Total Matches per Year
matches_per_year = df['Year'].value_counts().sort_index()

plt.figure()
plt.plot(matches_per_year.index, matches_per_year.values)
plt.title("Total Matches per Year")
plt.xlabel("Year")
plt.ylabel("Number of Matches")
plt.show()


# Most Winning Teams
top_winners = df['Winner'].value_counts().head(10)

plt.figure()
plt.bar(top_winners.index, top_winners.values)
plt.xticks(rotation=45)
plt.title("Top 10 Winning Teams")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.tight_layout()
plt.show()


# Matches per Ground
top_grounds = df['Ground'].value_counts().head(10)

plt.figure()
plt.bar(top_grounds.index, top_grounds.values)
plt.xticks(rotation=45)
plt.title("Top 10 Grounds by Matches")
plt.xlabel("Ground")
plt.ylabel("Matches Hosted")
plt.tight_layout()
plt.show()


# =====================================================
#  2 BIVARIATE ANALYSIS
# =====================================================

# Team1 vs Wins (Which team wins more when playing as Team1)
team1_wins = df[df['Team1'] == df['Winner']]['Team1'].value_counts().head(10)

plt.figure()
plt.bar(team1_wins.index, team1_wins.values)
plt.xticks(rotation=45)
plt.title("Top Wins as Team1")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.tight_layout()
plt.show()


# Team2 vs Wins
team2_wins = df[df['Team2'] == df['Winner']]['Team2'].value_counts().head(10)

plt.figure()
plt.bar(team2_wins.index, team2_wins.values)
plt.xticks(rotation=45)
plt.title("Top Wins as Team2")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.tight_layout()
plt.show()


# =====================================================
#  3 MULTIVARIATE ANALYSIS
# =====================================================

# Year vs Wins of Top Team
top_team = df['Winner'].value_counts().idxmax()
team_year_wins = df[df['Winner'] == top_team]['Year'].value_counts().sort_index()

plt.figure()
plt.plot(team_year_wins.index, team_year_wins.values)
plt.title(f"{top_team} Wins Over Years")
plt.xlabel("Year")
plt.ylabel("Wins")
plt.show()


# Ground vs Wins of Top Team
ground_wins = df[df['Winner'] == top_team]['Ground'].value_counts().head(10)

plt.figure()
plt.bar(ground_wins.index, ground_wins.values)
plt.xticks(rotation=45)
plt.title(f"{top_team} Wins by Ground")
plt.xlabel("Ground")
plt.ylabel("Wins")
plt.tight_layout()
plt.show()


print("Complete T20 Match Data Visualization Completed ")
```
